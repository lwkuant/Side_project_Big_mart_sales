#### Fill the NAs for Item_Weight

```python
### Fill the NAs based on the identical items
print('Original NAs:')
print(np.sum(df['Item_Weight'].isnull()))

group_item = df.groupby(['Item_Identifier'])
weight_table = dict(group_item['Item_Weight'].mean())

df['Item_Weight'][df['Item_Weight'].isnull()] = df['Item_Identifier'][df['Item_Weight'].isnull()].apply(lambda x: weight_table[x])

print('NAs after transformation:')
print(np.sum(df['Item_Weight'].isnull()))
```

```python
### Fill the NAs based on the identical Item Type (average)
print('Original NAs:')
print(np.sum(df['Item_Weight'].isnull()))

group_type = df.groupby(['Item_Type'])
weight_table = dict(group_type['Item_Weight'].mean())

df['Item_Weight'][df['Item_Weight'].isnull()] = df['Item_Type'][df['Item_Weight'].isnull()].apply(lambda x: weight_table[x])

print('NAs after transformation:')
print(np.sum(df['Item_Weight'].isnull()))
```

## Wrapper function

```python
### plot the relationship between numerical variables
def plot_num_relationship(df, x, y):
    import seaborn as sns
    plt.figure(figsize=[10, 6])
    axes = sns.regplot(x=x, y=y, data=df)
    
### plot the relationship between categorical variables and the target variable
def plot_cat_tar_relationship(df, x, y, orient='v'):
    import seaborn as sns 
    plt.figure(figsize=[10, 6])
    axes = sns.boxplot(x=x, y=y, data=df, orient=orient)
    
### calculate the anova value between categorical variables and target variable 
def cal_anova(df, y, x='Item_Outlet_Sales'):
    from scipy.stats import f_oneway
    vectors = []
    for cat in np.unique(df[y].values):
        vectors.append(df[x].values[df[y].values == cat])
    
    return f_oneway(*vectors)

### calculate the chi squared test of independence between categorical variables
def cal_chi2(df, col_name=[]):
    from scipy.stats import chi2_contingency
    
    chi_mat = np.zeros([len(col_name), len(col_name)])
    
    for i in range(len(col_name)):
        for j in range(len(col_name)):
            if i == j:
                chi_mat[i, j] = 0
            else:
                chi_mat[i, j] = (1 - chi2_contingency(pd.crosstab(df[col_name[i]], df[col_name[j]]))[1])
    
    chi_mat = pd.DataFrame(chi_mat, columns=col_name, index=col_name)
    
    return chi_mat

### Coonstruct the one hot encoding 
def label_encoder(df, col):
    from sklearn import preprocessing
    le = preprocessing.LabelEncoder()
    le.fit(df[col].values)
    
    return le

def one_hot(df, col, label_encoder):
    from sklearn import preprocessing
    vec = label_encoder.transform(df[col].values).reshape([len(df[col]), 1])
    ohc = preprocessing.OneHotEncoder(sparse=False, n_values=len(label_encoder.classes_))
    ohc_df = ohc.fit_transform(vec)
    
    ohc_df = pd.DataFrame(ohc_df, columns=label_encoder.classes_, index=range(len(df[col])))
    return ohc_df

### Transform the test ready for modeling 
def model_process(df, df_tr, num_col=['Item_Weight', 'Item_Visibility', 'Item_MRP'], cat_col=['Outlet_Type', 'Item_Type']):
    
    group_item = df.groupby(['Item_Identifier'])
    weight_table = dict(group_item['Item_Weight'].mean())
    df['Item_Weight'][df['Item_Weight'].isnull()] = df['Item_Identifier'][df['Item_Weight'].isnull()].apply(lambda x: weight_table[x])
    
    group_type = df.groupby(['Item_Type'])
    weight_table = dict(group_type['Item_Weight'].mean())
    df['Item_Weight'][df['Item_Weight'].isnull()] = df['Item_Type'][df['Item_Weight'].isnull()].apply(lambda x: weight_table[x])

    size_table = {'OUT010': 'Small', 'OUT045': 'Small', 'OUT017': 'Small'}
    df['Outlet_Size'][df['Outlet_Size'].isnull()] = df['Outlet_Identifier'][df['Outlet_Size'].isnull()].apply(lambda x: size_table[x])
    
    #df['Item_Outlet_Sales_log10'] = np.log10(df['Item_Outlet_Sales'].values+1)
    
    ### New feature
    #df['Item_Identifier_abbrev'] = df['Item_Identifier'].apply(lambda x: x[:2])
    
    def label_encoder(df, col):
        from sklearn import preprocessing
        le = preprocessing.LabelEncoder()
        le.fit(df[col].values)

        return le

    def one_hot(df, col, label_encoder):
        from sklearn import preprocessing
        vec = label_encoder.transform(df[col].values).reshape([len(df[col]), 1])
        ohc = preprocessing.OneHotEncoder(sparse=False, n_values=len(label_encoder.classes_))
        ohc_df = ohc.fit_transform(vec)

        ohc_df = pd.DataFrame(ohc_df, columns=label_encoder.classes_, index=range(len(df[col])))
        return ohc_df
    
    
    
    X = df.ix[:, num_col]
    #from sklearn.preprocessing import StandardScaler
    #for i in num_col:
        #scaler = StandardScaler()
        #scaler.fit(df_tr[i])
        #X[i] = scaler.transform(X[i])
    
    for i in cat_col:
        X = pd.concat([X, one_hot(df, i, label_encoder(df_tr, i))], axis=1)
        
    y = df['Item_Outlet_Sales']
    
    return X, y
    
```
